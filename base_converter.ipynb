```python
import gradio as gr

def base_converter(numero, base):
    try:
        if base == "Binário":
            decimal = int(numero, 2)
        elif base == "Octal":
            decimal = int(numero, 8)
        elif base == "Decimal":
            decimal = int(numero)
        elif base == "Hexadecimal":
            decimal = int(numero, 16)

        binario = bin(decimal)[2:]
        octal = oct(decimal)[2:]
        hexadecimal = hex(decimal)[2:].upper()

        output = f"<p><strong>Decimal:</strong> {decimal}</p>" \
                 f"<p><strong>Binário:</strong> {binario}</p>" \
                 f"<p><strong>Octal:</strong> {octal}</p>" \
                 f"<p><strong>Hexadecimal:</strong> {hexadecimal}</p>"
        
        return output
    except ValueError:
        return "Número inválido"

interface = gr.Interface(fn=base_converter, 
                         inputs=["text", gr.inputs.Radio(["Binário", "Octal", "Decimal", "Hexadecimal"], label="Base")], 
                         outputs="html",
                         title="Calculadora Conversora de Bases",
                         layout="vertical",
                         theme="soft",
                         examples=[
                             ["1101", "Binário"],
                             ["75", "Octal"],
                             ["127", "Decimal"],
                             ["1A3", "Hexadecimal"]
                         ],
                         examples_column_names=["Número", "Base"],
                         article="pt",
                         language="portuguese"
)

interface.launch(share=True)
```
